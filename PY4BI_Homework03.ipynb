### Задача 1.2

Создайте новый признак `delta_renovated`, который будет содержать разницу в годах между годом реновации дома (`yr_renovated`) и годом постройки дома (`yr_built`). Если реновации дома не было (год реновации равен 0), то в новом признаке должно быть значение 0.

```python
# Создаем признак delta_renovated
data['delta_renovated'] = data['yr_renovated'] - data['yr_built']
data.loc[data['yr_renovated'] == 0, 'delta_renovated'] = 0

# Выводим первые 5 строк для проверки
data.head()
```

### Задача 1.3

Создайте два новых признака: `year_sold` и `month_sold`, которые будут содержать год и месяц продажи соответственно, извлеченные из признака `date`.


```python
# Преобразуем столбец date в формат datetime
data['date'] = pd.to_datetime(data['date'])

# Создаем признак year_sold
data['year_sold'] = data['date'].dt.year

# Создаем признак month_sold
data['month_sold'] = data['date'].dt.month

# Выводим первые 5 строк для проверки
data[['date', 'year_sold', 'month_sold']].head()
```

Задача 3.5
Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома.

```python
data.pivot_table(index='condition', columns='grade', values='price', aggfunc='median')
```

Эта сводная таблица показывает медианную стоимость домов в зависимости от их состояния (condition) и оценки дома (grade).

Давайте проанализируем эту таблицу:

1. Дома с наивысшей оценкой (13) и в отличном состоянии (4) имеют самую высокую медианную стоимость — 5,750,000. 
2. Дома с наименьшей оценкой (1) и в плохом состоянии (1) имеют наименьшую медианную стоимость — 142,000.
3. Для домов с оценкой 1 и состоянием 2, 3, 4, 5 цены отсутствуют, что указывает на отсутствие таких домов в данных или на то, что такие дома не продавались.
4. С увеличением оценки дома его стоимость, как правило, увеличивается, что логично, так как лучший рейтинг указывает на лучшее качество или расположение дома.
5. Внутри одного и того же рейтинга дома в лучшем состоянии обычно стоят дороже.

На основе этих данных можно сделать вывод, чем выше оценка и состояние дома, тем выше его стоимость.
